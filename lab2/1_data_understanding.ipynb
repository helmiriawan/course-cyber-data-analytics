# Data Understanding

In this notebook, we will focus only on data understanding. It covers all activities that can help us to become familiar with the data, obtain initial insights, and discover issues in the data.

First, we need to import a few libraries that will be used in this notebook. After that, we will load the data and check some of the very first records. Note that there are two sets of data used in this notebook, which are the normal and attack dataset.

```python
# Import library
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Configuration
file_directory = "dataset" + "/"
normal_file = file_directory + "swat_normal_dataset.csv"
attack_file = file_directory + "swat_attack_dataset.csv"
```

```python
# Load input file
normal_data = pd.read_csv(normal_file)
attack_data = pd.read_csv(attack_file)
```

```python
# Show sample data
normal_data.head()
```

Next, we will examine and describe the data in terms of the format, the quantity, and the variables available in the data.

```python
# Check data type of all columns in normal data
normal_data.info()
```

```python
# Check data type of all columns in attack data
attack_data.info()
```

```python
# Generate descriptive statistics of normal data
normal_data.describe()
```

```python
# Generate descriptive statistics of attack data
attack_data.describe()
```

Here we found that there are 496,800 and 449,919 records in the normal and attack data respectively. There are 51 variables in both sets, and no missing values are found. Most of the variables are numerical, except two variables: "Timestamp", which provides the information about the time when the records are taken, and "Status", which provides the information about the status of the testbed. Note that all data is recorded once every second.

Furthermore, only five variables will be used in this lab, which are FIT101, LIT101, MV101, P101, and P102. Based on the [documentation](https://link.springer.com/chapter/10.1007/978-3-319-71368-7_8), those five variables are located in one of the chemical tanks in the testbed. Next, we will explore those variables by using data visualization.

```python
# Change data type of 'Timestamp' and 'Status'
for data in [normal_data, attack_data]:
    data['Timestamp'] = pd.to_datetime(
        data.Timestamp, 
        errors='coerce', 
        dayfirst=True
    )
    data['Status'] = data['Status'].astype('category')
```

```python
# Define the target variables
variables = ['FIT101', 'LIT101', 'MV101', 'P101', 'P102']
```

```python
# Plot the normal data
for variable in variables:
    ax = normal_data.plot(x='Timestamp', y=variable, figsize=(15, 4))
    ax.legend(loc='upper right');
```

From the figures above, we can see that there is something weird on December 22th. Based on the [documentation](https://link.springer.com/chapter/10.1007/978-3-319-71368-7_8), it is found that the testbed was in the empty state when the data collection process started. We might need to exclude this period since we are only interested in the normal condition for the modeling phase.

```python
# Plot the first 10 hours of normal data
for variable in variables:
    selected_data = normal_data[normal_data['Timestamp'] < '2015-12-23 01:23:59']
    ax = selected_data.plot(x='Timestamp', y=variable, figsize=(15, 4), color='orange')
    ax.legend(loc='upper right');
```

Next, we will take a closer look at the normal data and see how the patterns in the data look like.

```python
# Plot sample normal data
for variable in variables:
    selected_data = normal_data[
        (normal_data['Timestamp'] >= '2015-12-25 12:00:00')
        & (normal_data['Timestamp'] <= '2015-12-25 23:23:59')
    ]
    ax = selected_data.plot(x='Timestamp', y=variable, figsize=(15, 4))
    ax.legend(loc='upper right');
```

After we are done with the normal data, we will do similar processes to examine the attack data.

```python
# Plot the attack data
for variable in variables:
    ax = attack_data.plot(x='Timestamp', y=variable, figsize=(15, 4))
    ax.legend(loc='upper right');
```

```python
# Plot sample attack data
for variable in variables:
    selected_data = attack_data[
        (attack_data['Timestamp'] >= '2015-12-28 12:00:00')
        & (attack_data['Timestamp'] <= '2015-12-28 23:23:59')
    ]
    ax = selected_data.plot(x='Timestamp', y=variable, figsize=(15, 4), color='green')
    ax.legend(loc='upper right');
```
